### C5 (21)

Будем считать, что выбросами считаются наблюдения, которые отличаются от среднего арифметического на 0.5 стандартных отклонений и более. Определите, сколько дней НЕ считаются выбросами по признаку temperature. Ответ округлите до двух знаков. 

Пример ответа: 100.00

```python
# 1. Считаем границы
lower_border = data_c['temperature'].mean() - 0.5 * data_c['temperature'].std()
upper_border = data_c['temperature'].mean() + 0.5 * data_c['temperature'].std()

# 2. Фильтрация - удаляем выбросы
data_c_filtered = data_c[(data_c['temperature'] > lower_border) & (data_c['temperature'] < upper_border)]

print(f'Число выбросов по температуре - {data_c_filtered.shape[0]}')
```

### C7 (23)
Постройте модель линейной регрессии, которая показывает, как температура воздуха (**temperature**) влияет на количество ДТП с пешеходами (**pedestrian_incidents**). В качестве ответа укажите значение свободного коэффициента. Ответ округлите до двух знаков. 

Пример ответа: 0.00

```python
X = sm.add_constant(data_c['temperature'])
Y = data_c['pedestrian_incidents']

model = sm.OLS(Y, X).fit()

model.params.round(2)
```

### C8 (24)
Постройте модель линейной регрессии, которая показывает как температура воздуха (**temperature**) оказывает влияние на количество ДТП с пешеходами (**pedestrian_incidents**). Укажите в качестве ответа среднюю абсолютную ошибку (MAE) модели. Ответ округлите до двух знаков.

Пример ответа: 0.00

```python
Y_pred = model.predict(X)
mae = (Y - Y_pred).abs().mean()
print(f'MAE = {mae:.2f}')
```
