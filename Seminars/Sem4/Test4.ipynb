Добавьте в таблицу новый признак `coded`, который будет представлять собой кодированные значения признака `categories`. Признак `coded` должен содержать то, сколько раз строка Food  встретилась в значение `categories`. В ответ введите количество ресторанов, где Food встретился 1 раз и меньше.

|categories	| coded |
| --------- | ----- |
|Mexican Restaurant	| 0 |
|Fast Food	|  1 |
|Fast Food and Food	|  2 |

Пример ответа: 100

```python
data['coded'] = data['categories'].apply(lambda x: x.count('Food'))
data[data['coded'] <= 1].shape
```

Признак `coded` должен содержать то, сколько раз строка `Restaurant`  встретилась в значение `categories`. В ответ введите количество ресторанов, где `Restaurant` встретился более 1 раза.


|categories	| coded |

```python
data['coded'] = data['categories'].apply(lambda x: x.count('Restaurant'))
data[data['coded'] > 1].shape
```

Добавьте в таблицу новый признак `coded`, который будет представлять собой кодированные значения признака `categories`. Кодировку проведите следующим образом: если значение `categories` содержит строку Sandwich – 1, иначе – 0. Посчитайте сколько ресторанов содержат Sandwich в описании.

```python
data['coded'] = data['categories'].apply(lambda x: 'Sandwich' in x)
data[data['coded'] == 1].shape
```
