## B2
Графики рассеяния

Ход решения:
- График зависимости двух признаков - график рассеяния. Для его построения пользуемся `.plot(feature1, feature2, kind='scatter')`
- Каждая точка - одно наблюдение
- Если линейной зависимости нет, точки расположены на графике рандомно
- Если линейная зависимость есть, то можем представить "линию", аппроксимирующую значения обоих признаков
    - Положительная зависимость - с увеличением значений одного признака, увеличиваются значения другого признака (угол наклона прямой положительный)
    - Отрицательная зависимость - с увеличением значений одного признака, уменьшаются значения другого признака (угол наклона прямой отрицательный)
